```python
%load_ext autoreload
%autoreload 2
from solardatatools import DataHandler
from solardatatools.dataio import get_pvdaq_data
```

```python
import dask

@dask.delayed
def run_pipeline(year):
    df = get_pvdaq_data(sysid=34, year=year, api_key='DEMO_KEY')
    dh = DataHandler(df)
    dh.run_pipeline(power_col='ac_power')
    return dh.report(return_values=True)
```

```python
results = []
for i in range(3):
    results.append(run_pipeline(i+2011))
dask.compute(results)
```
